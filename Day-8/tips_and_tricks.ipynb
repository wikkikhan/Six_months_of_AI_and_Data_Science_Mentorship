# 3- How to rename columns?

```python
df = pd.DataFrame({'A col': [1, 2, 3, 6, 7, 8, 9, 10], 'B col': [4, 5, 6, 8, 9, 10, 11, 12]})
df
```

```python
df.rename(columns={'A col': 'col_a', 'B col': 'col_b'}, inplace=True)
df
```

```python
# Another way to rename columns
df.columns=['col_aa', 'col_bb']
df
```

```python
# to replace any character, string
df.columns=df.columns.str.replace('_', ' ')
df
```

```python
df.columns=df.columns.str.replace(' ', '_')
df
```

```python
# Adding prefix to columns
df = df.add_prefix('Numerical_')
df
```

```python
# Adding suffix to columns
df = df.add_suffix('_add')
df
```

# 16- Splitting a string into multiple columns

```python
# import libraries
import pandas as pd

df = pd.DataFrame({'name': ['Waqar Ahmad', 'Sana Waqar', 'Maryam Waqar', 'Ahmad khan'],
                   'location': ['Taxila, Pakistan', 'Gujrat, Pakistan', 'Lahore, Pakistan', 'Pakistan, Pakistan']})
df
```

```python
# spliting a column into two columns
df.name.str.split(' ', expand=True)
```

```python
# Adding those splits into new columns
df[["first_name", "last_name"]] = df.name.str.split(' ', expand=True)
df
```

```python
df[["city", "country"]] = df.location.str.split(",", expand=True)
df
```

```python
# Refine data manipulation
df = df[["first_name", "last_name", "city", "country"]]
df
```
